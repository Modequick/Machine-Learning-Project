```python
import random 
import matplotlib.pyplot as plt 
from statistics import mean 
```

Let say you want to have to know what is the probability you win when you play a 2 dice game. For every round, you will ned to pay \$1, and every time 2 die has match number, you will obtain \$4. What is the probability that you will winn ? 


We will be using monte carlo simulation to know this is. The higher number of round, will be represent more accurate true probability.


```python
#Simple Monte Carlo for Dice 
def roll_dice() : 
  die_1 = random.randrange(1,6)
  die_2 = random.randrange(1,6)

  if die_1 == die_2 :
    same_num = True
  else : 
    same_num = False 
  return same_num 
```

```python
#initial_input

num_simulation = 1000
round_simulation = 1000
opening_balance = 1000 
round_pay = 1 
if_win = 4

#tracking
ending_balance = []
winning_probability = []

```

```python
#creating the simulation
def simulation_dice():
  balance = opening_balance 
  win = if_win
  round_pay = 1
  container = []
  num_wins = 0
  for i in range(round_simulation):
    result = roll_dice()
    if result == True :
      balance += if_win
      num_wins += 1 
      container.append(balance)
    else :
      balance -= round_pay
      container.append(balance)
  wins_percentage = num_wins  / round_simulation

  return container,wins_percentage,balance
```

```python
#The plot for each simulation
win_container = []
for i in range (num_simulation):
  plt.plot(simulation_dice()[0])
  win_container.insert(i,simulation_dice()[-2])

 
print("The average win is :",mean(win_container))
```

```python
#the probability of ending balance >=1000
counter = 0 
for i in range(num_simulation):
  if simulation_dice()[-1] >1000 : 
    counter += 1 
print(counter * 100 / num_simulation ,"%")
```
